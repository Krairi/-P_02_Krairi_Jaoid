## DEMARCHE METHODOLOGIQUE DE L'ANALYSE DE DONNEES

### Supprimer des variables

En regardant les différentes variables ci-dessus, je vais tout d'abord supprimers les 2 id(id et id_emplacement) car il n'apporte rien à mon analyse, de plus je vais supprimer les variables suppérieur à 50% de valeurs manquantes (variete, complement_addresse,numero).

```python
df = df.drop(columns=['id','type_emplacement','complement_addresse','numero','id_emplacement','variete'])
```

### Couper notre dataframe en 2 parties ((Categorique : Qualitative) et (Numérique : Quantitative))

cat_data: Categorique et num_data: Numerique

```python
cat_data=[]
num_data=[]

for i,c in enumerate(df.dtypes):
    if c==object:
        cat_data.append(df.iloc[:,i])
    else:
        num_data.append(df.iloc[:,i]) 
cat_data=pd.DataFrame(cat_data).transpose()
num_data=pd.DataFrame(num_data).transpose()
```

Cette condition me permet de traiter de manière séparer les variables qualitative dite 'categorique' et les variables quantitative dite 'numerique'

### Remplacer les valeurs manquantes Categorique

Pour les variables qualitative on va remplacer les valeurs manquantes par les valeurs qui se répétent le plus souvent.

```python
cat_data=cat_data.apply(lambda x:x.fillna(x.value_counts().index[0]))
cat_data.isnull().sum()
```

Nous constatons qu'il n'y a plus aucune valeur qualitative manquante.

### Remplacer les valeurs manquantes Numerique

Pour les variables quantitative on va remplacer les valeurs manquantes par la médiane.

```python
num_data = num_data.fillna(num_data.median())
num_data.isnull().sum()
```

Nous constatons qu'il n'y a plus aucune valeur quantitative manquante, nèanmois il faudra aller plus loin au niveau des 2 variables quantitative 'circonference_cm' et 'hauteur_m' sachant que les deux autres variables sont des données géographique. La latitude correspond à la variable 'geo_point_2d_b' et la longitude correspond à la variable 'geo_point_2d_a'. De plus la variable 'remarquable' m’indique deux valeurs ‘0’ entretien non réalisé et ‘1’ entretien réalisé.

### Remplacer les valeurs aberrantes

#### Une valeur aberrante qu'est que c'est?

Une valeur aberrante est une valeur extrème ou une anomalie qui peut être décrite comme une données qui s'écarte trop du reste des observations, c'est à dire quelle se distingue nettement des autres.

#### Détecter les valeurs aberrante de la variable 'circonference_cm'

```python
sb.boxplot(x=df['circonference_cm'])
```

Le diagramme en moustache détecte un grand nombre de valeur aberrante celà ce distingue par une droite vertical et des points espacé.

#### Utilisation de la méthode interquartile  sur la variable 'circonference_cm'

```python
Q1=df['circonference_cm'].quantile(0.25)
Q3=df['circonference_cm'].quantile(0.75)
IQR=Q3-Q1
print(Q1)
print(Q3)
print(IQR)
Lower_Whisker = Q1 - 1.5*IQR
Upper_Whisker = Q3 + 1.5*IQR
print(Lower_Whisker, Upper_Whisker)
```

Cette méthode me permet de relever les points de données qui tombe en dehors de 1,5 fois d’une plage interquartile au-dessus du 3e quartile (Q3= 115.0) et en dessous du 1er quartile (Q1=30.0)

#### Traitement des valeurs aberrante 'circonference_cm'

La mèthoque que j'utilise ci-dessous est de définir une variable nommé mask intégrant les valeurs non aberrante, ensuite redéfinir la variable 'circonférence_cm' en lui imputant le quartile en pourcentage par tatonement sur les valeurs aberrante.

```python
mask = (df['circonference_cm']>0) & (df['circonference_cm']<Upper_Whisker)
df['circonference_cm'] = df['circonference_cm'].where(mask, df.loc[mask, 'circonference_cm'].quantile(0.80))
```

```python
sb.boxplot(x=df['circonference_cm'])
```

Le diagramme moustache nous montre que la variable 'circonference_cm' est totalement nettoyer, plus de présence de valeur aberrante.

Cette méthode me permet de relever les points de données qui tombe en dehors de 1,5 fois d’une plage interquartile au-dessus du 3e quartile (Q3= 12.0) et en dessous du 1er quartile (Q1=5.0)

#### Traitement des valeurs aberrante 'hauteur_m'

La mèthoque que j'utilise ci-dessous est de définir une variable nommé mask intégrant les valeurs non aberrante, ensuite redéfinir la variable 'hauteur_m' en lui imputant le quartile en pourcentage par tatonement sur les valeurs aberrante.

```python
mask = (df['hauteur_m']>0) & (df['hauteur_m']<Upper_Whisker)
df['hauteur_m'] = df['hauteur_m'].where(mask, df.loc[mask, 'hauteur_m'].quantile(0.20))
```

### Analysez la variable quantitative "circonference_cm"  et une qualitative "domanialite" 

```python
X = "domanialite"
Y = "circonference_cm" 
sous_echantillon = df[df["circonference_cm"] > 0].copy()
```

```python
modalites = sous_echantillon[X].unique()
groupes = []
for m in modalites:
    groupes.append(sous_echantillon[sous_echantillon[X]==m][Y])
    
medianprops = {'color':"black"}
meanprops = {'marker':'o', 'markeredgecolor':'black',
            'markerfacecolor':'firebrick'}
    
plt.boxplot(groupes, labels=modalites, showfliers=False, medianprops=medianprops, 
            vert=False, patch_artist=True, showmeans=True, meanprops=meanprops)
plt.show()
```

Les points rouges au milieu de chaque boîte à moustache représentent la moyenne des valeurs de circonférence en cm. On voit ici que les circonférences sont très différents d'une catégorie à l'autre. Par exemple, les circonférences des arbres dans le cimetière sont plus élevés et moins dispersés que les autres sauf le DASES.

### Analyse de 2 variables qualitatives "domanialite" et "stade_developpement"

```python
X = "domanialite"
Y = "stade_developpement"

c = df[[X,Y]].pivot_table(index=X,columns=Y,aggfunc=len)
cont = c.copy()

tx = df[X].value_counts()
ty = df[Y].value_counts()

cont.loc[:,"Total"] = tx
cont.loc["total",:] = ty
cont.loc["total","Total"] = len(df)
```

```python
tx = pd.DataFrame(tx)
ty = pd.DataFrame(ty)
tx.columns = ["foo"]
ty.columns = ["foo"]
n = len(df)
indep = tx.dot(ty.T) / n

c = c.fillna(0)
mesure = (c-indep)**2/indep
xi_n = mesure.sum().sum()
sb.heatmap(mesure/xi_n,annot=c)
plt.title('domanialite/stade_developpement Arbre')
plt.show()
```

En regardant ces cases foncées, on apprend que les arbres Mature sont souvent Alignement, que les arbres Jeune Adulte sont souvent DASCO, que les arbres Jeune sont souvent DJS et que les arbres Adulte  sont souvent DFPE.
